```python
name=[]
movie=[]

while(True):
    c=input("are want enter another user y or n:\n")
    if(c=="y"):
    
        while(True):
            a=input("enter name of user :\n")
            name.append(a)
            t=[]
            print("enter movies (**press n after done**)\n")
            while(True):
                b=input()
                if(b!="n"):
                    t.append(b)
                else:
                    break
            movie.append(t)
            break
    if(c=="n"):
        break
print(name)
print(movie)

re=list(zip(name,movie))
print(re)
```

```python
s=set()
for i in re:
    for j in i[1]:
         s.add(j)
print(s)
```

```python
l=[]
for i in range(len(re)):
    l.extend(re[i][1])
print(l)
```

```python
for i in s:
    if(l.count(i)==1):
        print(i)
```
