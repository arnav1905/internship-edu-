```python
name=[]
prod=[]

while(True):
    c=input("are want enter another customer y or n:\n")
    if(c=="y"):
    
        while(True):
            a=input("enter name of customer :\n")
            name.append(a)
            t=[]
            print("enter products (**press n after done**)\n")
            while(True):
                b=input()
                if(b!="n"):
                    t.append(b)
                else:
                    break
            prod.append(t)
            break
    if(c=="n"):
        break
print(name)
print(prod)

re=list(zip(name,prod))
print(re)
```

```python
s=set()
for i in re:
    for j in i[1]:
        s.add(j)
print(len(s))
```

```python
for i in range(len(re)):
    if("laptop" in re[i][1] and "headphone" in re[i][1]):
        print(re[i][0])
```

```python
x=[]
for i in prod:
    x.extend(i)
#print(x)
m=0
for i in x:
    s=x.count(i)
    if(m<s):
        m=s
s1=set()
for i in x:
    if(x.count(i)==m):
        s1.add(i)
print(s1)
```
