```python
# 8. Simulate do-while loop to check whether the entered number is prime or not.
n=int(input("enter number"))
x=2
while True:
    if(n<2):
        print("not prime")
        break
    if(n%x==0):
         print("not prime")
         break
    x=x+1
    if(x>=int(n**0.5)+1):
        print("not")
        break
        
    
```
