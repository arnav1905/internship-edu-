```python
# 4 Create two functions where both use a global variable to calculate and store 
#cumulative total sales

total= 0

def fun1(n):
    global total
    total = total + n

def fun2(n):
    global total
    total = total + n

fun1(50)
fun2(30)


print("Cumulative Total:", total)
```

```python
# 5 Write a nested function where the inner function accesses an enclosing variable 
#from the outer function
def out():
    x=25
    def inner():
        print(x)
    inner()
out()
```

```python
# 6 Demonstrate the use of nonlocal keyword by modifying an enclosing variable inside
#an inner function
def out():
    x=25
    def inner():
        nonlocal x
        x=65
        print(x)
    inner()
    print(x)
out()
```

```python
# 16 Write a function with variable positional arguments that finds the maximum
#value from all numbers passed
def demo(*n):
    m = n[0]

    for i in n:
        if i > m:
            m = i

    return m


print(demo(1,2,3,4,5))

```

```python
# 23 Write a program where an outer function accepts employee details and inner
#function calculates bonus based on salary
def out(**d):
    print("Employee Details:")
    for k, v in d.items():
        print(f"{k}: {v}")

    def inner():
        s = d.get("salary")  
        b = s * 0.2              
        print(f"Bonus: {b}")

    inner()
out(name="Arnav", department="IT", salary=50000)

```

```python
# 24 Demonstrate how a nonlocal variable can be used to maintain the count of how
#many times a nested function has been called
def out():
    c = 0  # enclosing variable

    def inner():
        nonlocal  c
        c += 1
        print(f"Function called {c} times")
    inner()
    inner()
    inner()

out()
```

```python
# 26 Create a nested function example where inner function modifies enclosing
#variable to accumulate running total of expenses
def out():
    t = 0 

    def inner(a):
        nonlocal t
        t += a
        print(f"Added {a}, Total Expense: {t}")

    inner(100)
    inner(50)
    
out()

```

```python
# 29 Write a function with both *args and **kwargs to simulate order details 
#containing items and their prices
def demo(*q, **p):
    print("Order Items:")
    for i in q:
        print(f" {i}")

    print("\nItem Prices:")
    for k, v in p.items():
        print(f"{k}: {v}")

demo("Laptop", "Mouse", "Keyboard", Laptop=1200, Mouse=25, Keyboard=45)
```
