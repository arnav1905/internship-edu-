```python
dw1=[]
dw2=[]

while(True):
    c=input("are you want add product in first y or n:\ ")
    if(c=="y"):
    
        while(True):
            t=[]
            a=input("enter product id :\n")
            b=input("enter product name:\n ")
            c=input("enter product quantity\n")
            t.append(a)
            t.append(b)
            t.append(c)
            break
        dw1.append(t)
        
    if(c=="n"):
        break
print(dw1)

while(True):
    c=input("are you want add product in second y or n:\ ")
    if(c=="y"):
    
        while(True):
            t=[]
            a=input("enter product id :\n")
            b=input("enter product name:\n ")
            c=input("enter product quantity\n")
            t.append(a)
            t.append(b)
            t.append(c)
            break
        dw2.append(t)
        
    if(c=="n"):
        break
print(dw2)
```

```python
for i in range(len(dw1)):
    for j in range(len(dw2)):
        if(dw1[i][1]==dw2[j][1]):
            print(dw1[i][1])
```
