```python
name=[]
course=[]

while(True):
    c=input("are you want to add student y or n:\n")
    if(c=="y"):
    
        while(True):
            a=input("enter name of student :\n")
            name.append(a)
            t=[]
            print("enter courses (**press n after done**)\n")
            while(True):
                b=input()
                if(b!="n"):
                    t.append(b)
                else:
                    break
            course.append(t)
            break
    if(c=="n"):
        break
#print(name)
#print(course)
nt=tuple(name)
ct=tuple(course)
print(nt)
print(ct)
re=list(zip(name,course))
#print(re)
```

```python
for i in range(len(re)):
    if("ds" in re[i][1] and "ai" in re[i][1]):
        print(re[i][0])
```
