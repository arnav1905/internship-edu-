```python
attendance = [set() for _ in range(5)]

for day in range(5):
    print(f"Enter day {day+1} attendance (**press n when done**)")
    while True:
        a = input()
        if a == "n":
            break
        attendance[day].add(a)

s1, s2, s3, s4, s5 = attendance

print("Day 1:", s1)
print("Day 2:", s2)
print("Day 3:", s3)
print("Day 4:", s4)
print("Day 5:", s5)
```
