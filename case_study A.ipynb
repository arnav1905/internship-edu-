```python
name=[]
skill=[]

while(True):
    c=input("are want enter another employee y or n:\n")
    if(c=="y"):
    
        while(True):
            a=input("enter name of employee :\n")
            name.append(a)
            t=[]
            print("enter skills (**press n after done**)\n")
            while(True):
                b=input()
                if(b!="n"):
                    t.append(b)
                else:
                    break
            skill.append(t)
            break
    if(c=="n"):
        break
print(name)
print(skill)

re=list(zip(name,skill))
print(re)
```
